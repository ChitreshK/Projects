```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
match_data=pd.read_csv(r"IPL Matches 2008-2020.csv")
match_data.head()
```

```python
each_ball_data=pd.read_csv(r"IPL Ball-by-Ball 2008-2020.csv")
each_ball_data.head()
```

```python
# extract dataframe which contain info of each batsman in single match  
single_match_data = each_ball_data[each_ball_data['id']==335986]
```

```python
# create list of each batsman from given match  
all_batsman=list(single_match_data['batsman'].unique())
all_batsman
```

```python
single_match_data.head()
```

```python
# create dictionary with team name as keys 
run_of_each_player = dict.fromkeys(all_batsman)

for player in all_batsman:
    total_run=sum(single_match_data[single_match_data['batsman']==player]['batsman_runs'])
    
    # total run scored by each batsman
    x=single_match_data[single_match_data['batsman']==player]['dismissal_kind'].unique()
    print(f"{player}{x}")
    #remove nan from list of dissmisal type 
    my_list=list(x)
  
    new_list = []
    for item in my_list:
        if str(item) != 'nan':
            new_list.append(item)
    
    dismissal_type=new_list
    
    # check condition when dismissial is not given 
    if len(dismissal_type)==0:
        dismissal_type=["not out"]
    temp_list=[total_run,dismissal_type]
    run_of_each_player[player]=temp_list
```

```python
df=pd.DataFrame(columns={"batsman","dismissal"})
for key in run_of_each_player.keys():
    row=len(df.index)
    df.loc[row,"batsman"]=key
    df.loc[row,"dismissal"]=run_of_each_player[key][1][0]
df.to_csv("C:/Users/navya/Desktop/DEproject/questions_csv/question2_dismissal.csv",index=False)
```
